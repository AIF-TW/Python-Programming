## 變數的運算符號



Python 對於各種變數支援不同的運算符號集。整數、浮點數及複數支援一般四則運算、次方等，還有邏輯符號(and，or，not)及比較符號(>=，==，<=，!=)，而字串則支援串接(+)，重複(*)，也支援比較符號。

### Arithmetic Operators

| Symbol | Task Performed |
|----|---|
| +  | Addition |
| -  | Subtraction |
| /  | division |
| %  | mod |
| *  | multiplication |
| //  | floor division |
| **  | to the power of |

```python
a, b = 10, 3

print(f'{a}  +  {b} = {a + b}')
print(f'{a}  -  {b} = {a - b}')
print(f'{a}  *  {b} = {a * b}')
print(f'{a}  /  {b} = {a / b}') # 一般除法，就算是兩個整數相除，也會將結果自動轉為浮點數

# Floor division，只留下商
print(f'{a} //  {b} = {a // b}')

# 取餘數
print(f'{a} % {b} = {a % b}')

# 指數寫法，例如10 的 3 次方
print(f'{a} **  {b} = {a ** b}')
```

```python
s1 = 'Hello'
s2 = 'World'

# 字串的 + 符號為串接意思
S = s1 + ' ' + s2
print(S)
# 字串的 * 符號為重複意思
S = (s1+' ') * 5 + (s2+' ') * 3
print(S)
```

### Comparison Operators

| Symbol | Task Performed |
|----|---|
| == | True, if it is equal |
| !=  | True, if not equal to |
| < | less than |
| > | greater than |
| <=  | less than or equal to |
| >=  | greater than or equal to |

```python
a = 3.1
b = 3

# 整數與符點數的大小比較相當直觀，但注意複數型態是無法做大小比較的。
print(f'{a}  > {b}? {a > b}')
print(f'{a}  < {b}? {a < b}')
print(f'{a} == {b}? {a == b}')
print(f'{a} >= {b}? {a >= b}')
print(f'{a} <= {b}? {a <= b}')
print(f'{a} != {b}? {a != b}')
```

```python
a = 'abcde'
b = 'abcdf'

# 字串的大小比較是依照字典序，想像它們會在字典出現的先後順序，先出現者較小。
print(f'{a}  > {b}? {a > b}')
print(f'{a}  < {b}? {a < b}')
print(f'{a} == {b}? {a == b}')
print(f'{a} >= {b}? {a >= b}')
print(f'{a} <= {b}? {a <= b}')
print(f'{a} != {b}? {a != b}')
```
